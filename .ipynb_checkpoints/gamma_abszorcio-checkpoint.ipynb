```python
Elméleti Összefoglaló:
Gamma abszorcio:
Miközben a γ - sugárzás áthalad egy anyagon, intenzitása gyengül. A gyengülés
mértéke függ a sugárzás fotonenergiájától, az anyag vastagságától, minőségétől és
sűrűségétől.

FOTOEFFEKTUS: A γ -foton teljes energiája a
kilépési munkává (A), és a kilökött fotoelektron
mozgási energiájává alakul.
    
Compton-SZÓRÁS: A g -foton energiája a
kilépési munkává (A), valamint a kilökött
Compton-elektron mozgási energiájává, ill. a szórt
és gyengített energiájú Compton-foton energiájává
alakul.

PÁRKÉPZŐDÉS: Csak 1.022 MeV-nál nagyobb
γ - foton energiáknál fordulhat elő. Ekkor a foton
eltűnik és energiájából egy elektron és egy pozitron
keletkezik. Később a pozitron lefékeződve és az
anyag egy elektronjával egyesülve megsemmisül
(annihilálódik), miközben az impulzusmegmaradás
törvénye miatt, két 511 keV-os γ -fotonná alakul.

LINEÁRIS GYENGÍTÉSI EGYÜTTHATÓ (μ): Az anyag
minőségétől és sűrűségétől, valamint a sugárzás energiájától
függő együttható, amely azt jellemzi, milyen mértékben nyeli
el az adott anyag a sugárzást. Mértékegysége: 1/cm (1/m).

FELEZŐ RÉTEGVASTAGSÁG (D): Az a rétegvastagság,
amely J0 belépő intenzitást a felére csökkenti.
    
TIZEDELŐ RÉTEGVASTAGSÁG (x1/10): Az a rétegvastagság,
amely a J0 belépő intenzitást a tizedére csökkenti.
    
TÖMEGGYENGÍTÉSI EGYÜTTHATÓ (μm): Az anyag
minőségétől, valamint a sugárzás energiájától (de az anyag
sűrűségétől nem) függő együttható, amely azt jellemzi, milyen
mértékben nyeli el az adott anyag a sugárzást.
Mértékegysége: cm2/g (m2/kg).

FELÜLETI SŰRŰSÉG (xm): Az abszorbens 1 cm2 felülete
mögött elhelyezkedő tömeg. Mértékegysége g/cm2 (kg/m2).

FELEZŐTÖMEG (Dm): A felező rétegvastagság és a sűrűség
szorzata: Dm = D * r . Mértékegysége g/cm2 (kg/m2).
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Mérési adatok betöltése
data = pd.read_excel('gamma.xlsx')
data.set_index('anyag', inplace=True)
print(data)
```

```python
## Eredmények táblázata
output = pd.read_excel('data.xlsx')
output.set_index('abszorbens', inplace=True)
print(output)

## háttérsugárzás (J0) [J]
j0 = data.loc['háttérsugárzása', 'átlag']
print('j0=' ,j0)
```

```python
## 1-2 feladat
## Féllogaritmukus skálán ábrázolva az adott anyaggal mért átlag intenzitás eredmények a rétegvastagság függvényében
## ezután egyenest illesztek az ábrára amivel meg lehet határozni az N0 majd az N0 / 2 értékeket, melyből a D kiszámítható
def plot_intensity_and_calc_D(anyag):
    x = (data.loc[anyag, 'mennyiség'])
    y = (data.loc[anyag, 'átlag'] - j0)
    slope, intercept = np.polyfit(list(x), list(y), 1)  ## egyenes egyenletének paraméterei
    approx_line = slope * (x) + intercept  ## illesztett egyenes
    plt.semilogy(x, y, label='mérési adatok')
    plt.semilogy(x, approx_line, label='illesztett egyenes')
    plt.title(anyag)
    plt.xlabel('rétegvastagság cm')
    plt.ylabel('Intenzitás J')

    ## calc D
    N0 = intercept;
    half_N0 = N0 / 2
    D = (half_N0 - intercept) / slope  ## D
    print('D =',D, 'cm')
    output.loc[anyag, 'D'] = D
    plt.scatter(0, N0, label='(0, N0)')
    plt.scatter(D, half_N0, label='(D, half_N0)')
    plt.legend()
    plt.show()

plot_intensity_and_calc_D('ólom')
plot_intensity_and_calc_D('alumínium')
plot_intensity_and_calc_D('vas')
plot_intensity_and_calc_D('ólomgumi')
print(output)
```

```python
## 3-4 feladat
## lineáris gyengítési együttható és tizedelő rétegvastagság számítása és az eredmények beírása az output táblázatba
## D = ln2 / u
## X 1/10 = 3.33D
for index, row in output.iterrows():
    output.loc[index, 'u'] = np.log(2) / row['D']
    output.loc[index, 'X 1/10'] = 3.33 * row['D']
print(output)
```

```python
## 5-6 feladat
## tömeggyengítési egyötthatók és felezési tömegek
## um = u / p
## Dm = ln2 / um
for index, row in output.iterrows():
    output.loc[index, 'um'] = row['u'] / row['P']
    output.loc[index, 'Dm'] = np.log(2) / row['um']

print(output)
```
